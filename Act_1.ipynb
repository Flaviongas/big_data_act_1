```python
def agregar_campeon():
    global df 
    Nombre_Campeon = input("Escribe el nombre del Campeón: ")
    Rol = input("Escribe el rol del Campeón (e.g., Top, Jungle, Mid, ADC, Support): ")
    Vida_base = input("Escribe los puntos de vida del Campeón: ")
    Mana_base = input("Escribe el daño de ataque del Campeón: ")
    Armadura_base = input("Escribe la armadura base: ")
    Daño_ataque_base = input("Escribe el daño de ataque base: ")
    Oro = input("Escribe el oro: ")
    
    campeon = pd.DataFrame({
        'Nombre Campeon': [Nombre_Campeon],
        'Rol': [Rol],
        'Vida base': [Vida_base],
        'Mana base': [Mana_base],
        'Armadura base': [Armadura_base],
        'Daño ataque base': [Daño_ataque_base],
        'Oro': [Oro],
        })
    ##aqui lo que quiero hacer es añadir lo de arriba al dataframe original , pero el append no funciona entonces uso el concat
    df = pd.concat([df,campeon], ignore_index=True)
```

```python
def buscar_campeon():
    name = input("Escribe el nombre del Campeón: ")
    resultados = df[df['Nombre Campeon'].str.contains(name, case=False, na=False)]
    
    if not resultados.empty:
        print(f"Se encontraron {len(resultados)} resultados:")
        print(resultados)
    else:
        print("No se encontró el Campeón.")
```

```python
def modificar_campeon():
    global df
    nombre_original = input("Escribe el nombre del Campeón que deseas modificar: ")
    campeon_existente = df[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False)]
    
    if campeon_existente.empty:
        print(f"No se encontró el Campeón '{nombre_original}'.")
        return
    
    print("Datos actuales del campeón:")
    print(campeon_existente)
    ["Nombre Campeon","Rol", "Vida base","Mana base" , "Armadura base" , "Daño ataque base" ," Oro"]

    nuevo_nombre = input("Escribe el nuevo nombre del Campeon: ")
    nuevo_rol = input("Escribe el nuevo rol del Campeon: ")
    nuevo_hp = input("Escribe los nuevos puntos de vida del Campeon: ")
    nuevo_ap = input("Escribe los nuevos puntos de mana base del Campeon: ")
    nuevo_armor = input("Escribe los nuevos puntos armadura del Campeon: ")
    nuevo_attack = input("Escribe los nuevos puntos de ataque del Campeon: ")
    nuevo_Oro = input("Escribe los nuevos puntos de oro del Campeon: ")
    
    if nuevo_nombre:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Nombre Campeon'] = nuevo_nombre
    if nuevo_rol:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Rol'] = nuevo_rol
    if nuevo_hp:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Vida base'] = nuevo_hp
    if nuevo_ap:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Mana base'] = nuevo_ap
    if nuevo_armor:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Armadura base'] = nuevo_armor
    if nuevo_attack:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Daño ataque base'] = nuevo_attack

    if nuevo_Oro:
            df.loc[df['Nombre Campeon'].str.contains(nombre_original, case=False, na=False), 'Oro'] = nuevo_Oro
        
```
